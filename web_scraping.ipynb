```python
final = pd.DataFrame()
for j in range(1,11):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j) , headers=headers).text
    soup = BeautifulSoup(webpage , 'lxml')
    company =  soup.find_all('div' , class_='companyCardWrapper')

    cname = []
    reviews = []
    rating = []
    jobs = []
    for i in company:
        cname.append(i.find('h2' , class_='companyCardWrapper__companyName').text.strip())
        reviews.append(i.find('span' , class_='companyCardWrapper__companyRatingCount').text.strip())
        rating.append(i.find('div' , class_='rating_text rating_text--md').text.strip())
        jobs.append(i.find_all('span' , class_='companyCardWrapper__ActionCount')[3].text.strip())

    d = {'name' : cname , 'reviews' : reviews , 'ratings' : rating , 'jobs' : jobs}
    df = pd.DataFrame(d)
    final = pd.concat([final, df], ignore_index=True)
    
```
